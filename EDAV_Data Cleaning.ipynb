```python
# drop columns which are all nans or have over 99% nans
data.dropna(axis=1, how = "all", inplace=True)
missing = ['Intersection Street 1', 'Intersection Street 2']
data.drop(missing, axis=1, inplace=True)
data.shape
```

```python
# drop columns with the constant value
data = data.loc[:,data.apply(pd.Series.nunique) != 1]
data.shape
```

```python
# drop columns which are highly correlated with others
redundant = ['Park Borough','X Coordinate (State Plane)', 'Y Coordinate (State Plane)', 'Location']
data.drop(redundant, axis=1, inplace=True)
data.shape
```

```python
# At this time, we finish the basic cleaning on the data, then we will explore the cleaned data.
data.to_csv('data_cleaned.csv', index=False)
```
